```python
import os
import platform
import matplotlib.font_manager as font_manager
import matplotlib.pyplot as plt
import pandas as pd
from typing import Optional
import logging as lg

def set_plt() -> None:
    ## set plt minus font
    plt.rcParams["axes.unicode_minus"] = False

    ## set font
    system_name = platform.system()
    if system_name == "Windows":
        # Windows
        plt.rc("font", family="Malgun Gothic")
    elif system_name == "Darwin":
        # Mac
        plt.rc("font", family="AppleGothic")
    elif system_name == "Linux":
        # Linux
        path = "/usr/share/fonts/truetype/nanum/NanumMyeongjo.ttf"
        font_name = font_manager.FontProperties(fname=path, size=12)
        plt.rc("font", family=font_name)
    else:
        print("Not support")

set_plt()

df1 = pd.read_csv("C:\\Users\\dkq50\\OneDrive\\바탕 화면\\bipolar_disorder.csv", sep = ',', encoding = 'utf-8', header = [0,1])
df1
#환자수만 추출
#df2 = df1.loc[:,df1.columns.get_level_values(1) == [('성별','연령구분10세','환자수')]]
```

```python
df1.columns
col1 = df1.loc[:,[('구분','성별'),('진료년월','연령구분10세')]]
col2 = df1.iloc[:,df1.columns.get_level_values(1) == '환자수']

result = pd.concat([col1,col2],axis=1)
result
```

```python
#성별 소계 

cp = result.copy() 
idx = cp.set_index(('구분','성별'))
idx

m = cp.iloc[1, 2:]

f = cp.iloc[11,2:]


#man = m.to_frame()
#fm = f.to_frame()

man = m.T
female = f.T

print(man.size)
print(female.size)
#sex = pd.concat([m, f], axis = 1)
#print(type(sex))

man = pd.to_numeric(man)
female = pd.to_numeric(female)
```

```python
x_axis = man.index.map(lambda x:x[0])
y_axis = man.values.tolist()

x_axis1 = female.index.map(lambda x:x[0])
y_axis1 = female.values.tolist()

plt.figure(figsize=(35,8))

plt.plot(x_axis, y_axis, label = '남성')
plt.plot(x_axis1, y_axis1, label = '여성')

plt.title('성별 조울증 진단 환자 수 추이(2019~2021)')

plt.xlabel('진료년월')
plt.ylabel('성별 환자 수(명)')
plt.legend()


plt.show()
```
